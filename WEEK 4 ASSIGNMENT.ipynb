#Data Visualization

##Univariate data analysis

```python
df.sample(7)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Visualization of the value count for each specie

sns.countplot(data =df, x= 'species')
```

```python
#Visualize all individual numerical columns using histplot

sns.histplot(df['sepal_length_cm'])
```

```python
#Create a 2x2 subplot displaying the individual numerical columns
fig, axes = plt.subplots(2,2, figsize=(12,8))

#Plot the first histplot on row 0 and column 0 index
sns.histplot((df['sepal_length_cm']), ax = axes[0,0], kde = True);
axes[0,0].set_title('sepal_length_cm')

#Plot the second histplot on row 0 and column 1 index
sns.histplot((df['sepal_width_cm']), ax = axes[0,1], kde = True);
axes[0,1].set_title('sepal_width_cm')

#Plot the third histplot on row 1 and column 0 index
sns.histplot((df['petal_length_cm']), ax = axes[1,0], kde = True);
axes[1,0].set_title('petal_length_cm')

#Plot the fourth histplot on row 1 and column 1 index
sns.histplot((df['petal_width_cm']), ax = axes[1,1], kde = True);
axes[1,1].set_title('petal_length_cm')

#adjust spacing
plt.tight_layout()

#show the plot
plt.show()
```

#Outlier Detection using Box Plot

```python
sns.boxplot(data = df, x= 'sepal_length_cm')
```

```python
sns.boxplot(data=df, y='species', x ='sepal_length_cm')
```

```python
# Creation of a 2x2 subplot grid - 2 rows and 2 columns
fig, axes = plt.subplots(2, 2, figsize=(12,8))

#plot the first displot on row 0 and column 0 index 
sns.boxplot(data = df , y = 'species', x= 'sepal_length_cm', ax=axes[0,0])
axes[0,0].set_title('Sepal Length')

#plot the second displot on row 0 and column 1 index 
sns.boxplot(data = df , y = 'species', x= 'sepal_width_cm', ax=axes[0,1])
axes[0,1].set_title('Sepal width')

#plot the third displot on row 1 and column 0 index 
sns.boxplot(data = df , y = 'species', x= 'petal_length_cm', ax=axes[1,0])
axes[1,0].set_title('Petal Length')

#plot the fourth displot on row 1 and column 1 index 
sns.boxplot(data = df , y = 'species', x= 'petal_width_cm', ax=axes[1,1])
axes[1,1].set_title('Petal Width')

plt.tight_layout()
plt.show()
```
