```python
Categories = ['Ant', 'Bird', 'Cat', 'Dog']
Values = [25, 40, 15, 30]

sns.barplot(x= Categories, y= Values,);

plt.title('Population of animals')
plt.xlabel('Animal')
plt.ylabel('number of each')
plt.show()
```

```python
data = [32, 45, 50, 28, 36, 42, 38, 25, 29, 52, 48, 60]
sns.histplot(x=data, bins=8, kde = True)

plt.title('Data points')
plt.xlabel('Data group')
plt.ylabel('Frequency')
plt.show()
```

#Pie Chart

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
Sizes =  [30, 25, 20, 15]

plt.pie(Sizes, labels = labels, autopct = '%1.2f%%')
plt.show()
```
